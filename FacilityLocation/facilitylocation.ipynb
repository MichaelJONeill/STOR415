```julia
# Code for this example is based on code from Laurent Lessard: https://laurentlessard.com/

using JuMP, HiGHS, Random

nF = 100  # number of facilities
nC = 100  # number of customers

# generate random problem instance
Random.seed!(1)
#f = rand(nF)
f = 0.5*ones(nF)
c = rand(nF,nC)

m = Model(HiGHS.Optimizer)

@variable(m, x[1:nF], Bin)         # binary constraint

@variable(m, y[1:nF,1:nC], Bin)     # binary constraint

@constraint(m, sumconstraint[j = 1:nC], sum( y[i,j] for i=1:nF ) == 1)

# choose one of these two constraints (they are equivalent)
#@constraint(m, cr[i=1:nF], sum( y[i,j] for j=1:nC ) <= nC*x[i])
@constraint(m, cr[i=1:nF,j=1:nC], y[i,j] <= x[i])

@objective(m, Min, sum(f[i]*x[i] for i in 1:nF) + sum(sum(c[i,j]*y[i,j] for i in 1:nF) for j in 1:nC))

optimize!(m)
```
